```python
X=news_train["text"]
Y=news_train["label"]

vectorizer=CountVectorizer()

vectorizer.fit(X)

transformed=vectorizer.transform(X)
transformed
```

```python
labels=(Y=='REAL')
labels=labels.astype(int)

labels
```

# classifier
